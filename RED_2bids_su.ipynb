```python
# Run DCM2BIDS for all RED paricipants on cluster
import subprocess

# to adjust which participants to run
# primarily for testing purposes to see this is working
start = 7
stop = 8

# Iterate over path names and subject number, calling cbu2bids at the command line
for n, participant in enumerate(sub_paths[start:stop], start=start):
    sbatch_cmd = "sbatch --mem-per-cpu=6gb --time=01:00:00 --job-name=RED2bids \
    --output /imaging/su01/RED/bids_slurm/sub-%03d-RED2bids.out" % (n+1)
   
    bids_cmd = " dcm2bids -d " + participant + '/*' + " -p %03d " % (n+1) + "-o " + out_dir + \
    " -c /imaging/su01/RED/config.json " + "-a pydeface"
   
    cmd = sbatch_cmd + bids_cmd
    cmd
   
    p = subprocess.run((cmd), shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
   
    print(p.args)
    print(p.stdout.decode())
    print(p.stderr.decode())
```

```python
# Update subject specific json files for the ANATOMICAL images
### Change to 3 digit subject IDs

import json
import os.path

for p in range(len(dir_list)): 
    
    # Open the json file for the FUNCTIONAL image and load the data
    try:
        with open ("/imaging/su01/RED/bids/sub-%03d/anat/sub-%03d_T1w.json" %(p+1, p+1), 'r') as anat:
            dat = json.load(anat)

        # Update values in the json dictionary
        dat.update({'PulseSequenceType': 'MPRAGE',
                     'PulseSequenceDetails': 'Standard Siemens Sequence distributed with VD13D Software',
                     'NonlinearGradientCorrection': False,
                     'ParallelAcquisitionTechnique': 'GRAPPA',
                     'PartialFourierDirection': 'j-',
                     'PhaseEncodingDirection': 'j-',
                     'InstitutionName': 'University of Cambridge', 
                     'InstitutionalDepartmentName':'Cognition and Brain Sciences Unit'})

        # Write the dictionary to the json file
        with open ("/imaging/su01/RED/bids/sub-%03d/anat/sub-%03d_T1w.json" %(p+1, p+1), 'w') as ana:
            json.dump(dat, ana, indent=4)
            
    except FileNotFoundError as e:
        # Subjects with re-runs were not found - print and save paths to a csv
        print("sub-%03d" %(p+1), e)
        with open('/imaging/su01/RED/bids/RED_T1_Reruns.csv','a') as file:
            file.write("sub-%03d " %(p+1) + str(e))
            file.write('\n')
        
        # Repeat json update for subjects with re-runs
        for n in range(3):
            try:
                with open ("/imaging/su01/RED/bids/sub-%03d/anat/sub-%03d_run-%02d_T1w.json" %(p+1, p+1, n+1), 'r') as anat:
                    dat = json.load(anat)

                # Update values in the json dictionary
                dat.update({'PulseSequenceType': 'MPRAGE',
                             'PulseSequenceDetails': 'Standard Siemens Sequence distributed with VD13D Software',
                             'NonlinearGradientCorrection': False,
                             'ParallelAcquisitionTechnique': 'GRAPPA',
                             'PartialFourierDirection': 'j-',
                             'PhaseEncodingDirection': 'j-',
                             'InstitutionName': 'University of Cambridge', 
                             'InstitutionalDepartmentName':'Cognition and Brain Sciences Unit'})

                # Write the dictionary to the json file
                with open ("/imaging/su01/RED/bids/sub-%03d/anat/sub-%03d_run-%02d_T1w.json" %(p+1, p+1, n+1), 'w') as ana:
                    json.dump(dat, ana, indent=4)
            except FileNotFoundError as fe:
                print(fe)
```
